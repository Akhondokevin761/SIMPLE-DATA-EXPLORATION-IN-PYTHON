```python
#Grouping the smartphones by RAM, then define an agg_function that will be used in finding the mean, sum, min, max and count in each ram of different variables as follows
agg_function = {
    'Sale Price':
    ['sum','mean','min','max','count'],
    'Mrp':['sum','mean','min','max','count'],
    'Discount Percentage':['sum','mean','min','max','count'],
    'Number Of Ratings':['sum','mean','min','max','count'],
    'Number Of Reviews':['sum','mean','min','max','count']
}
group1 = smart_data.groupby(['Ram']).agg(agg_function)
# Rename the columns for better readability
group1.columns = ['_'.join(col).strip() for col in group1.columns.values]
# Reset the index for a cleaner look
group1 = group1.reset_index()
print(group1)
```

```python
#Grouping the smartphones by Brand, then define an agg_function that will be used in finding the mean, sum, min, max and count in each ram of different variables as follows
agg_function = {
    'Sale Price':
    ['sum','mean','min','max','count'],
    'Mrp':['sum','mean','min','max','count'],
    'Discount Percentage':['sum','mean','min','max','count'],
    'Number Of Ratings':['sum','mean','min','max','count'],
    'Number Of Reviews':['sum','mean','min','max','count']
}
group2 = smart_data.groupby(['Brand']).agg(agg_function)#Grouping the smartphones by Brand, then define an agg_function that will be used in finding the mean, sum, min, max and count in each ram of different variables as follows
# Rename the columns for better readability
group2.columns = ['_'.join(col).strip() for col in group2.columns.values]
# Reset the index for a cleaner look
group2 = group2.reset_index()
print(group2)
```

```python
#We define a function that is called ratings that takes Star Ratings and assigns poor, good and best.
def Rating(x):
    if x<2.0:
        return 'Poor Performance '
    elif x>=2.0 and x<=3.8:
        return 'Average Performance'
    else:
        return 'Best Performance'
```

```python
#Using the above function, we create a new column called Performance and then apply the our rating function as follows
smart_data['Performance'] = smart_data['Star Rating'].apply(Rating)
smart_data.head(10)
```

```python
#We can now obtain the number of smartphones in the three categories by using a count function
Count_function = {
    'Sale Price':
    ['count','mean','sum']}
#Apply the function as follows
count_2 = smart_data.groupby(['Performance']).agg(Count_function)
count_2.head(5)
```
